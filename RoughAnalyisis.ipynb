```python
nyc = pd.read_csv('Directory_Of_Homeless_Population_By_Year.csv')
nyc.head()
```

```python
areas = nyc['Area'].unique()
plt.figure(figsize=(11, 6))

for area in areas:
    area_data = nyc[nyc['Area'] == area] 
    plt.plot(area_data['Year'], area_data['Homeless Estimates'], label = area)

plt.title('Homeless Count by Area Over the Years', fontsize = 16)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Homeless Count', fontsize = 14)
plt.xticks(nyc['Year'].unique(), fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(title = 'Area')
plt.grid(True, linestyle = '--')

plt.tight_layout()
plt.show()
```
